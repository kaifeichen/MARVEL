```python
import matplotlib.pyplot as plt
%matplotlib inline  
```

```python
dataFile = open("feature_and_ages.txt")
x = []
y = []
while(True):
    line1 = dataFile.readline()
    if not line1:
        break
    line2 = dataFile.readline()
    line3 = dataFile.readline()
    x.append(int(line1))
    y.append(float(line2))
#     print(str(int(line1)) + " " + str(float(line2)) + line3)
    
```

```python
# plt.yscale('log')
# plt.xscale('log')
plt.plot(x, y, 'ro')
plt.ylabel('mumber of edges in the image')
plt.xlabel('number of features')
```
